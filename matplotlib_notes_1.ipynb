Bar graph- Comparison plots:

```python
fruits = [1,2,3]
market1 = [10,60,40]
market2 = [12,40,30]
plt.bar(fruits, market1, width=0.2)
plt.bar(fruits, market2, width=0.2) # market2 is on top.
plt.legend(['market1', 'market2']) 
plt.show()
```

```python
fruits = [1,2,3]
market1 = [10,60,40]
market2 = [12,40,30]
plt.bar(fruits, market2, width=0.2)
plt.bar(fruits, market1, width=0.2) # now market1 is on top.
plt.legend(['Market2', 'Market1']) 
plt.show()
```

Data is overlapping. We want it side by side. Hence,

```python
fruits = [1,2,3]
market1 = [10,60,40]
market2 = [12,40,30]
plt.bar(fruits, market2, width=0.2)
plt.bar(fruits+0.2, market1, width=0.2) 
plt.legend(['Market1', 'Market2']) 
plt.show()
```

```python
# everything has to be np.array. Otherwise element wise operation won't happen.
fruits = np.array([1,2,3])
market1 = np.array([10,60,40])
market2 = np.array([12,40,30])
plt.bar(fruits, market2, width=0.2)
plt.bar(fruits+0.2, market1, width=0.2) # added .2 to fruits.
plt.legend(['Market1', 'Market2']) 
plt.show()
```

**Histogram:**

The hist() function returns:

n, bins, patches = plt.hist(R, nbins, normed=True)
n is an array of the bar heights, and bins is the array of bin boundaries.

```python
vals=[1,2,3,1,2,3,3,2,4,3,2,1,2,2,4]
plt.hist(vals) # shows the frequency of values.
plt.show() # 1 appears 3 times. 2 appears 6 times. 4 appears 2 times.
```

```python
#most frequently occuring value = 2 which is mode.
vals=[1,2,3,1,2,3,3,2,4,3,2,1,2,2,4]
plt.hist(vals) 
plt.show()
```

```python
vals=[1,2,3,1,2,3,3,2,4,3,2,1,2,2,4]
plt.hist(vals, bins=5) 
plt.show()
```

```python
vals=[1,2,3,1,2,3,3,2,4,3,2,1,2,2,4]
plt.hist(vals, bins=4) # create 4 equally spaced values. last value is included. 
# [1,1.8), [1.8, 2.5), [2.5, 3.2), [3.2, 4.0] - the height would depend on the number of values in these ranges.
# bins would be: [1, 1.8, 2.5, 3.2, 4.0]  
plt.show()
```

```python
# without giving bins
vals=[1,2,3,1,2,3,3,2,4,3,2,1,2,2,4,7]
plt.hist(vals)
plt.show()
```

```python
# with bins
vals=[1,2,3,1,2,3,3,2,4,3,2,1,2,2,4,7]
plt.hist(vals, bins=4) # x + 1 ie. 4 + 1 = 5 values
plt.show()
```
